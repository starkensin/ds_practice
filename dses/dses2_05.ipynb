```python
query = """
    select *
    from sqlite_master
    where 
        type='table'
"""
pd.read_sql(query,db)
```

```python
# inner join
# Customers <- Oders
query="""
    select * 
    from Customers C
    Join Orders O
    where C.CustomerID = O.CustomerID
"""
pd.read_sql(query,db)
```

```python
# inner join
# Customers <- Oders
query="""
    select * 
    from Customers C, Orders O
    where C.CustomerID = O.CustomerID
"""
pd.read_sql(query,db)
```

```python
# inner join
# Customers <- Oders
query="""
    select * 
    from Customers C
    Join Orders O
        On C.CustomerID = O.CustomerID
"""
pd.read_sql(query,db)
```

```python
# left join

query="""
    select C.ContactName, O.OrderID
    from Customers C
    left Join Orders O
        on C.CustomerID = O.CustomerID
"""
pd.read_sql(query,db)
```

```python
# self join

query="""
    select C1.CustomerName "CN1", C2.CustomerName "CN2", C1.City
    from Customers C1, Customers C2
    where
        C1.City=C2.City
        and C1.CustomerID!=C2.CustomerID
"""
pd.read_sql(query,db)
```
