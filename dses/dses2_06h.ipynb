```python
import requests
from bs4 import BeautifulSoup as btfs
```

```python
url="https://rottentomatoes.com"
response=requests.get(url)
response
```

```python
html=response.text
soup=btfs(html,"html.parser")
top=soup.find('div',attrs={'id':'homepage-top-box-office'})
top
```

```python
table_title=top.find('h2').text
table_title
```

```python
data_table=top.find('table')
data_table
```

```python
data_table=top.find_all('tr')
data_table
```

```python
for item in data_table:
    a_list=item.find_all('a')
    try:
        score=a_list[0].find('span',attrs={"class":'tMeterScore'}).text
    except:
        score='not yet'
    movie_name=a_list[1].text
    sales=a_list[2].text.strip()
    print(score,movie_name,sales,"\n")
```
