```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Problem 2

x = np.linspace(0, 100, 100)
plt.plot(x,2*x)
plt.xlabel("x")
plt.ylabel("y")
k=plt.axes([0.25, 0.5, 0.15, 0.15], xlabel='x', ylabel='y', xlim=[0,100], ylim=[0,200])
k.plot(x,2*x)
```
