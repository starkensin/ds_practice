```python
# How to merge with (1) pandas
query ="""
    select *
    from Customers
    ;
"""
customers_df=pd.read_sql(query,db)

query ="""
    select *
    from Orders
    ;
"""
orders_df=pd.read_sql(query,db)
```

```python
orders_df.head()
```

```python
customers_df.head()
```

```python
integrated_df=orders_df.merge(customers_df,on="CustomerID")
integrated_df.head()
```

```python
integrated_df=orders_df.merge(customers_df,on="CustomerID")[["OrderID","CustomerID","ContactName","Address"]]
integrated_df.head()
```

```python
# How to merge with (2) Sql

query = """
    select O.OrderID, C.CustomerName
    from Customers C, Orders O
    where
        C.CustomerID == O.CustomerID;
"""
pd.read_sql(query, db).head()
```

```python
# How to merge with (3) Sql Join
query = """
    select *
    from Customers C
    join Orders O On C.CustomerID=O.CustomerID;
"""
pd.read_sql(query, db).head()
```

```python
# For a month
order_df["Month"]=order_df["OrderDate"].apply(lambda x : str(x)[:7])
order_df.head()
```

```python
order_groups=order_df.groupby("Month")
order_count = pd.DataFrame([
    {
        "OrderDate": Month,
        "Count": len(order_groups.get_group(Month)), 
    } for Month in order_df["Month"].unique()
])
order_count.head()
```
