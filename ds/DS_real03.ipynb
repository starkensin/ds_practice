```python
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data=pd.read_csv("health_2016.csv",nrows=1000)
print(data.shape)
data.head()
```

```python
mpl.rc('font',family='Malgun Gothic')
```

```python
sns.countplot(data=data,x="성별코드")
```

```python
plt.figure(figsize=[7,4])
sns.countplot(data=data,x="성별코드")
```

```python
figure,((ax1,ax2))=plt.subplots(nrows=1,ncols=2)
figure.set_size_inches(24,4)
sns.countplot(data=data,x="성별코드",ax=ax1)
sns.countplot(data=data,x="시도코드",ax=ax2)
```

```python
sns.countplot(data=data,x="연령대코드(5세단위)",hue="성별코드")
```

```python
sns.countplot(data=data,x="연령대코드(5세단위)", hue="성별코드" ,color='b')
```

```python
palette=sns.color_palette('bright')
sns.countplot(data=data,x="연령대코드(5세단위)",palette=palette)
```

```python
plt.figure(figsize=[4,7])
sns.countplot(data=data,y="시도코드")
```

### barplot

```python
plt.figure(figsize=[15,4])
sns.barplot(data=data,x='연령대코드(5세단위)',y='수축기혈압')
```

```python
plt.figure(figsize=[15,4])
sns.barplot(data=data,x='연령대코드(5세단위)',y='수축기혈압',ci='sd')
```

```python
import numpy as np
plt.figure(figsize=[15,4])
sns.barplot(data=data,x='연령대코드(5세단위)',y='수축기혈압',ci='sd', estimator=np.sum)
```

```python
plt.figure(figsize=[5,8])
sns.barplot(data=data,x='수축기혈압',y='연령대코드(5세단위)',orient='h')
```

```python
mpl.rc('axes',unicode_minus=False)
```

```python
plt.figure(figsize=[15,4])
data["수축기혈압(copy)"]=data['수축기혈압']-data['수축기혈압'].mean()
sns.barplot(data=data,x='연령대코드(5세단위)',y='수축기혈압(copy)')
```
