### Merge

```python
query='''
SELECT *
FROM 'transaction'
'''
transaction=pd.read_sql(query,connection)
transaction
```

```python
query='''
SELECT *
FROM 'user'
'''
user=pd.read_sql(query,connection)
user
```

```python
pd.merge(transaction,user,how='inner')
```
